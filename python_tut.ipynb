
# Variables

Variables in Python are much the same as other coding languages.
Python is a dynamically-typed language so the exact type of the variable is determined at runtime.
The basic Data Types are:
* Numbers
* String
* List
* Tuple
* Dictionary


## Rules for Variable Names

* Must start with a letter or an underscore
* Cannot start with a number
* Can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Are case-sensitive

<div class="alert alert-block alert-info">
e.g.
To assign the variable age the value of 28
or assign red to car (notice that strings require quotes(single or double))
</div>

```python
age = 28
car = ('red')
print(age)
print(car)
```

<div class="alert alert-block alert-info">
To view the data type of a variable use the type() function.
</div>

```python
print(type(age)) # this should return <class 'int'>
print(type(car)) # this should return <class 'str'>
```

# Basic Data Types
The following examples will use a print command to display the result although this is not required in Jupyter notebook or Colab it makes the code easier to use in other IDE's like VScode.

## Numbers

### Integers or Int

```python
print(3)
print(type(3))
```

### Float

```python
print(4.5)
print(type(4.5))
```

### Complex


```python
print(1+7j)
print(type(1+7j))
```

## Strings

Strings in python are enclosed in either single or double quotes.

```python
house = ("A place to call home")
print(house)
print(type(house))
print(len(house))
```

Python does not have a character type instead it uses a string of length 1

```python
print(type('a'))
print(len('a'))
```
